## LOADING DATAS

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
climate_data= pd.read_csv("../Mobile-money-financial-inclusion-analyzer/data/Agrofood_co2_emission.csv")
climate_data.head(20)
```

# Extracting data of Kenya and Uganda

```python
# East Africa rows, only Kenya and Uganda columns
countries1 = climate_data[(climate_data["Area"] == "Kenya") | (climate_data["Area"] == "Uganda")]
countries1
```

# Extracting data of Ethiopia and Rwanda

```python

countries2 = climate_data[(climate_data["Area"] == "Ethiopia") | (climate_data["Area"] == "Rwanda")]
countries2.head(15)
```

## EXTRACTING DRC AND DJIBOUTI

```python

countries3 = climate_data[(climate_data["Area"] == "Democratic Republic of the Congo") | (climate_data["Area"] == "Djibouti")]
countries3.head(15)
```

## EXTRACTING SOMALI AND ERITREA

```python

countries4 = climate_data[(climate_data["Area"] == "Somali") | (climate_data["Area"] == "Eritrea")]
countries4.head(15)
```

```python
countries5 = climate_data[(climate_data["Area"] == "Sudan") | (climate_data["Area"] == "South Sudan")]
countries5.head(15)
```

# Concatenation of data

```python
east_africa_data = pd.concat([countries1, countries2,countries3,countries4,countries5], axis=0, ignore_index=True)
east_africa_data
```

```python
 # for col in df_num.columns:
 #   plt.figure(figsize=(8, 6))
 #   sns.boxplot(x=df_num[col])
 #   plt.title(f'Box Plot of {col}')
 #   plt.show()
# Total number of numeric columns
import math
num_cols = len(df_num.columns)

# Grid layout: Adjust columns per row
cols_per_row = 3
rows = math.ceil(num_cols / cols_per_row)

# Create the subplots grid
fig, axes = plt.subplots(rows, cols_per_row, figsize=(cols_per_row * 5, rows * 4))
axes = axes.flatten()  # Flatten in case of multi-row grid

# Plot each boxplot
for i, col in enumerate(df_num.columns):
    sns.boxplot(x=df_num[col], ax=axes[i], color='blue')
    axes[i].set_title(f'Boxplot of {col}', fontsize=12)
    axes[i].set_xlabel('')
    axes[i].set_ylabel('Value')

# Hide any unused axes
for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])

# Overall layout tweaks
plt.suptitle("ð Boxplots of Numeric Features", fontsize=16, weight='bold', y=1.02)
plt.tight_layout()
plt.show()
```

```python
def get_whiskers(col):
    Q1 = col.quantile(0.25)
    Q3 = col.quantile(0.75)
    IQR = Q3 - Q1
    lower_whisker = Q1 - 1.5 * IQR
    upper_whisker = Q3 + 1.5 * IQR
    return lower_whisker,upper_whisker

outlier_indices = set()

for col in df_num.columns:
    lw,uw = get_whiskers(df_num[col])
    index = np.where((df_num[col] > uw) | (df_num[col] < lw))
    outlier_indices.update(index[0])

df_num.drop(index = list(outlier_indices),inplace = True)

# for col in df_num.columns:
#     plt.figure(figsize = (8,6))
#     sns.boxplot(x = df_num[col])
#     plt.title(f"Box plot of {col}")
#     plt.show()
# Total number of numeric columns
num_cols = len(df_num.columns)

# Grid layout: Adjust columns per row
cols_per_row = 3
rows = math.ceil(num_cols / cols_per_row)

# Create the subplots grid
fig, axes = plt.subplots(rows, cols_per_row, figsize=(cols_per_row * 5, rows * 4))
axes = axes.flatten()  # Flatten in case of multi-row grid

# Plot each boxplot
for i, col in enumerate(df_num.columns):
    sns.boxplot(x=df_num[col], ax=axes[i], color='blue')
    axes[i].set_title(f'Boxplot of {col}', fontsize=12)
    axes[i].set_xlabel('')
    axes[i].set_ylabel('Value')

# Hide any unused axes
for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])

# Overall layout tweaks
plt.suptitle("ð Boxplots of Numeric Features", fontsize=16, weight='bold', y=1.02)
plt.tight_layout()
plt.show()
```

## Feature scaling

```python
#implementation of minmax scaler
from sklearn.preprocessing import MinMaxScaler
mms = MinMaxScaler()
df_num_scaled = mms.fit_transform(df_num)
df_num_scaled#the result is an array.we need to convert to a dataframe
df_num = pd.DataFrame(df_num_scaled, columns=df_num.columns)#here we equate to df_num so as to update the dataframe
df_num.head(10)#the result is a dataframe with values between 0 and 1
```

```python
#import libraries
#determine x
#determine Y
#do train test split
#initiate linear regression
#do mean squared error,r2score
from sklearn import metrics as m
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
x = df_num.iloc[0:229, 0:26]  # First 20 rows, first 5 columns
y = df_num.total_emission
y

```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=1 )
```

```python
linreg=LinearRegression()
linreg.fit(x_train,y_train)
```

```python
print(linreg.intercept_)
```

```python
cof= (linreg.coef_) 
cof
```

```python
y_pred=linreg.predict(x_test)
```

```python
m.mean_squared_error(y_test,y_pred)
```

```python
np.sqrt(m.mean_squared_error(y_test,y_pred ))
```

```python
y_train_pred= linreg.predict(x_train) # training dataset(x_train and y_train)
print(np.sqrt(m.mean_squared_error(y_train,y_train_pred)))
print(m.r2_score(y_train, y_train_pred)) # accuracy on training dataset

y_test_pred= linreg.predict(x_test)
print(np.sqrt(m.mean_squared_error(y_test,y_test_pred))) 
print(m.r2_score(y_test, y_test_pred))
```

```python
from sklearn.linear_model import Ridge
from sklearn.linear_model import Lasso
from sklearn.metrics import r2_score

rr = Ridge(alpha=0.01)
rr.fit(x_train, y_train) #RIDGE REGRESSION

pred_train_rr= rr.predict(x_train) #PREDICTIONS ON TRAINING SET

print('MSE: TRAINING SET')
print(np.sqrt(m.mean_squared_error(y_train,pred_train_rr)))
print('SCORE: TRAINING SET')
print(r2_score(y_train, pred_train_rr))

pred_test_rr= rr.predict(x_test) # PREDICTIONS ON TESTING TEST

print('MSE: TESTING SET')
print(np.sqrt(m.mean_squared_error(y_test,pred_test_rr))) 
print('SCORE: TESTING SET')
print(r2_score(y_test, pred_test_rr))
```
