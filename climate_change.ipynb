## LOADING DATAS

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
climate_data= pd.read_csv("../Data/Agrofood_co2_emission.csv")
climate_data.head()
```

# Extracting data of Kenya and Uganda

```python
# East Africa rows, only Kenya and Uganda columns
countries1 = climate_data[(climate_data["Area"] == "Kenya") | (climate_data["Area"] == "Uganda")]
countries1
```

# Extracting data of Ethiopia and Rwanda

```python

countries2 = climate_data[(climate_data["Area"] == "Ethiopia") | (climate_data["Area"] == "Rwanda")]
countries2.head(15)
```

# Concatenation of data

```python
east_africa_data = pd.concat([countries1, countries2], axis=0, ignore_index=True)
east_africa_data
```

```python
east_africa_data.isnull()
```

```python
df_num = east_africa_data.select_dtypes(include='float')
df_num
```

```python
df_num.corr()
```

# Dropping columns

```python
east_africa_data.drop(['Drained organic soils (CO2)', 'Food Household Consumption','Forestland','Food Retail','Fertilizers Manufacturing','Fires in organic soils',], axis=1, inplace=True)
```

# Detection of outliers

```python
 for col in df_num.columns:
   plt.figure(figsize=(8, 6))
   sns.boxplot(x=df_num[col])
   plt.title(f'Box Plot of {col}')
   plt.show()
```

```python
def get_whiskers(col):
    Q1 = col.quantile(0.25)
    Q3 = col.quantile(0.75)
    IQR = Q3 - Q1
    lower_whisker = Q1 - 1.5 * IQR
    upper_whisker = Q3 + 1.5 * IQR
    return lower_whisker,upper_whisker

outlier_indices = set()

for col in df_num.columns:
    lw,uw = get_whiskers(df_num[col])
    index = np.where((df_num[col] > uw) | (df_num[col] < lw))
    outlier_indices.update(index[0])

df_num.drop(index = list(outlier_indices),inplace = True)

for col in df_num.columns:
    plt.figure(figsize = (8,6))
    sns.boxplot(x = df_num[col])
    plt.title(f"Box plot of {col}")
    plt.show()
```

```python
for col in df_num.columns:
    plt.figure(figsize=(8,6))
    sns.kdeplot(x = df_num[col])
    plt.title(f"KDE plot of {col}")
    plt.show()
```

```python
from sklearn.preprocessing import PowerTransformer
pt = PowerTransformer(method='yeo-johnson')
for col in df_num:
    reshaped = df_num[[col]].values
    transformed = pt.fit_transform(reshaped)
    
    df_num[col] = transformed 
```

```python
for col in df_num.columns:
    plt.figure(figsize=(8,6))
    sns.kdeplot(x = df_num[col])
    plt.title(f"KDE plot of {col}")
    plt.show()
```
